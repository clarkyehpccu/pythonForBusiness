```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10.0, 10.0, 500)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, '.--b')
plt.plot(x,y2, 'x--r')
```

# .--b   .為接點, -- 為繪線   b:Blue   /   

```python
import pandas as pd
index_x = x
Pie = pd.Series(y1, index=index_x)
Pie.plot(kind='bar')
```
